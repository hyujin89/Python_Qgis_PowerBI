## Mobile Map ##

- 좌표를 지도화하여 반경 설정과 클러스터를 통해 가독성을 높일 수 있음

- html파일은 크롬에서 열면되고 Documents 어플을 통해 저장한 파일별로 볼 수 있음

- 마커별 추가 정보(사진/데이터 등)를 넣을 수 있지만 보안상 별도 관리함이 좋을 것 같음


```python
import folium
from folium import Circle
import pandas as pd
from folium.plugins import MarkerCluster  # MarkerCluster import

# CSV 파일에서 노드 정보 읽어오기
nodes_df = pd.read_csv('node1.csv')
nodes_df2 = pd.read_csv('이마.csv')

# 지도의 중심 좌표와 확대 정도 설정
map_center = [37.5, 126.9]  # 노드 평균 좌표를 중심으로 설정
zoom_level = 12

# Folium 지도 생성 (OpenStreetMap 타일 사용)
m = folium.Map(location=map_center, zoom_start=zoom_level, tiles='OpenStreetMap') # 여러가지 스타일이 있음
marker_cluster = MarkerCluster().add_to(m)


layer_control = folium.LayerControl(position='topright')


# 각 노드에 대해 반경 1km의 Circle 마커 추가
for index, row in nodes_df.iterrows():
    node_coord = [row['y'], row['x']]
#     Circle(location=node_coord, radius=2000, color='blue', fill=True, fill_color='orange', fill_opacity=0.7).add_to(m)
#     Circle(location=node_coord, radius=1000, color='blue', fill=True, fill_color='orange', fill_opacity=0.7).add_to(m)
    folium.Marker(location=node_coord, popup=row['점포명'], icon=folium.Icon(color='orange', icon='star')).add_to(marker_cluster)  #클러스터로 보이게 할 경우

# 각 노드에 대해 반경 1km의 Circle 마커 추가 
for index, row in nodes_df2.iterrows():
    node_coord = [row['y'], row['x']]
    
    Circle(location=node_coord, radius=1000, color='blue', fill=True, fill_color='orange', fill_opacity=0.1).add_to(marker_cluster).add_to(m)  #radius=1000반경 설정
#     Circle(location=node_coord, radius=10000, color='blue', fill=True, fill_color='orange', fill_opacity=0.1).add_to(marker_cluster).add_to(m)
    folium.Marker(location=node_coord, radius=1000, popup=row['점포명'], icon=folium.Icon(color='Red', icon='info-sign',size=0.1)).add_to(marker_cluster)
    
    popup_content = f"<b>{row['name']}</b><br>Address: {row['address']}<br>Phone: {row['phone']}"
    
    popup=folium.Popup(popup_content, max_width=300)
    
    
# LayerControl 객체를 추가
layer_control.add_to(m)
# 지도를 출력
m
```

```python
# HTML 파일로 저장
m.save('Mobile Map.html')
```
